```python
import cv2
body_classifier=cv2.CascadeClassifier('C:/Users/mayank singh/Desktop/Opencv/haarcascade_frontalface_default.xml')

img=cv2.imread('C:/Users/mayank singh/Desktop/family11/tt100.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces=body_classifier.detectMultiScale(gray,1.3,5)

print(len(faces))
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(225,0,0),5) 
    
cv2.imshow('upper body',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
